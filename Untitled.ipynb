```python
import re

# CSV file load
df = pd.read_csv("spam.csv", encoding="latin1")

# "Message" column clean
df['Message'] = df['Message'].apply(lambda x: re.sub(r'[^A-Za-z0-9 ]', '', str(x)))  # special characters remove
```

```python
df['Message']
```

```python
df
```

```python
import nltk
from nltk.corpus import stopwords

# stopwords download (only once)
nltk.download('stopwords')


# English stopwords list
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # 1. Lowercase
    text = str(text).lower()

    # 2. Remove punctuation + special characters
    text = re.sub(r'[^a-z0-9 ]', '', text)

    # 3. Remove stopwords
    words = text.split()
    words = [w for w in words if w not in stop_words]

    # 4. Join words again (no spaces required?  if no space, remove below)
    text = " ".join(words)

    return text

# Apply on Message column
df['Message'] = df['Message'].apply(clean_text)
```

```python
df['Message']
```

```python
df.columns
```

```python
x=df['Message']
y=df['Category']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the vectorizer
vectorizer = TfidfVectorizer(max_features=5000) 

# Fit the vectorizer ONLY on the TRAINING data and transform both sets
X_train_vec = vectorizer.fit_transform(x_train)
X_test_vec = vectorizer.transform(x_test)

print(f"Text converted to numerical features. Training matrix shape: {X_train_vec.shape}")
print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
print(len(x))
print(len(y))
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
model.fit(X_train_vec,y_train)
```

```python
y_pred=model.predict(X_test_vec)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
cr=classification_report(y_test,y_pred)
```

```python

# The 'model' and 'vectorizer' variables are available from your previous steps

# --- 1. Save the Trained Model ---
# We save the MultinomialNB model object
model_filename = 'spam_classifier_model.pkl'
with open(model_filename, 'wb') as file:
    pickle.dump(model, file)
print(f" Trained Model saved as: {model_filename}")

# --- 2. Save the TF-IDF Vectorizer ---
# We save the vectorizer to ensure new text data is transformed identically
vectorizer_filename = 'tfidf_vectorizer.pkl'
with open(vectorizer_filename, 'wb') as file:
    pickle.dump(vectorizer, file)
print(f" TF-IDF Vectorizer saved as: {vectorizer_filename}")
```
